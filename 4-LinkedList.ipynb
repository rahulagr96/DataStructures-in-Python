```python
class Node:
    def __init__(self,val = None, n=None, p=None):
        self.val = val
        self.next = n
        self.prev = p
    def __str__(self):
        return ('('+str(self.val)+')')
```

```python
class LinkedList:
    def __init__(self, r=None):
        self.root = r
        self.size = 0

    def print_list(self):
        this_node = self.root
        while this_node is not None:
            print(this_node,end='->')
            this_node = this_node.next
        print("None")
    
    def insertFront(self, newData):
        nn = Node(newData)
        nn.next = self.root
        self.root = nn
        self.size += 1
    
    def insertEnd(self,newData):
        nn = Node(newData)
        this_node = self.root
        while this_node.next is not None:
            this_node = this_node.next
        this_node.next = nn
        self.size += 1
    
    def insertAfter(self, preNode, newData):
        nn = Node(newData)
        this_node = self.root
        while this_node.val == preNode.val:
            this_node = this_node.next
        this_node = this_node.next
        nn.next = this_node.next
        this_node.next = nn
        self.size += 1
    
    def insertBefore(self, preNode, newData):
        nn = Node(newData)
        this_node = self.root
        while this_node.val == preNode.val:
            this_node = this_node.next
        nn.next = this_node.next
        this_node.next = nn
        self.size += 1

    def findNode(self, item):
        this_node = self.root
        i = 1
        while this_node is not None:
            if this_node.val == item:
                return i
            else:
                this_node = this_node.next
            i += 1
        return None 

    def remove(self, item):
        back_node = None
        front_node = self.root
        while front_node is not None:
            if front_node.val == item:
                if back_node is not None:
                    back_node.next = front_node.next
                else:
                    self.root = front_node.next
                return True
            else:
                back_node = front_node
                front_node = front_node.next
        return False
```

## Circular Link List

```python
class CircularLinkedList:
    def __init__(self, r=None):
        self.root = r
        self.size = 0

    #inserts to front and if empty
    def insert(self, newData):
        if self.size == 0:
                self.root = Node(newData)
                self.root.next = self.root
        else:
            nn = Node(newData)
            nn.next = self.root.next
            self.root.next = nn
        self.size += 1
    
    def insertEnd(self,newData):
        nn = Node(newData)
        this_node = self.root
        while this_node.next is not self.root:
            this_node = this_node.next
        nn.next = this_node.next
        this_node.next = nn
        self.size += 1
    
    def insertAfter(self, preNode, newData):
        nn = Node(newData)
        this_node = self.root
        while this_node.val == preNode.val:
            this_node = this_node.next
        this_node = this_node.next
        nn.next = this_node.next
        this_node.next = nn
        self.size += 1
    
    def insertBefore(self, preNode, newData):
        nn = Node(newData)
        this_node = self.root
        while this_node.val == preNode.val:
            this_node = this_node.next
        nn.next = this_node.next
        this_node.next = nn
        self.size += 1

    def findNode(self, item):
        this_node = self.root
        i = 1
        while True:
            if this_node.val == item:
                return i
            elif this_node.next == self.root:
                return False
            this_node = this_node.next
            i += 1

    def remove(self, item):
        back_node = None
        front_node = self.root
        while True:
            if front_node.val == item:
                if back_node is not None:
                    back_node.next = front_node.next
                else:
                    while front_node.next != self.root:
                        front_node = front_node.next 
                    front_node.next = self.root.next
                    self.root = self.root.next
                self.size -= 1
                return True
            elif front_node.next == self.root:
                return False
            back_node = front_node
            front_node = front_node.next

    def print_list(self):
        this_node = self.root
        while this_node is not None:
            print(this_node,end='->')
            this_node = this_node.next
        print("None")
```
