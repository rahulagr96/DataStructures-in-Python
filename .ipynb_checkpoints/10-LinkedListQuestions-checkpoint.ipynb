## Insert on index using recursion

```python
class Node:
    def __init__(self,val = None, n=None):
        self.val = val
        self.next = n
    def __str__(self):
        return ('('+str(self.val)+')')
        
class LinkedList:
    def __init__(self, r=None):
        self.root = r
        self.size = 0

    def print_list(self):
        this_node = self.root
        while this_node is not None:
            print(this_node,end='->')
            this_node = this_node.next
        print("None")

    def insertFront(self, newData):
        nn = Node(newData)
        nn.next = self.root
        self.root = nn
        self.size += 1
    
    def __insertRec(self, val, index, node:Node):
        if index == 0:
            nn = Node(val, node)
            self.size += 1
            return nn
        node.next = self.__insertRec(val, index-1, node.next) 
        return node

    def insetUsingRec(self, val, index):
        self.root = self.__insertRec(val, index, self.root)
```

```python
lis = LinkedList()
lis.root = Node("Mon")
lis.print_list()
lis.insetUsingRec("tue",0)
lis.print_list()
lis.insetUsingRec("Wed",1)
lis.print_list()
lis.insetUsingRec("Wee",1)
lis.print_list()
lis.insetUsingRec("Sat",3)
lis.print_list()
lis.insetUsingRec("Sat",5)
lis.print_list()
```

## Delete common elements

```python
class Node:
    def __init__(self,val = None, n=None):
        self.val = val
        self.next = n
    def __str__(self):
        return ('('+str(self.val)+')')
        
class LinkedList:
    def __init__(self, r=None):
        self.root = r
        self.size = 0

    def addRoot(self, val):
        self.root = Node(val)
        self.size += 1

    def print_list(self):
        this_node = self.root
        while this_node is not None:
            print(this_node,end='->')
            this_node = this_node.next
        print("None")

    def insertFront(self, newData):
        nn = Node(newData)
        nn.next = self.root
        self.root = nn
        self.size += 1
    
    def __insertRec(self, val, index, node:Node):
        if index == 0:
            nn = Node(val, node)
            self.size += 1
            return nn
        node.next = self.__insertRec(val, index-1, node.next) 
        return node

    def insetUsingRec(self, val, index):
        self.root = self.__insertRec(val, index, self.root)

    def removeDuplicates(self):
        node = self.root
        while node.next != None:
            if node.val == node.next.val:
                node.next = node.next.next
                self.size -= 1
            else:
                node = node.next
```

```python
lis = LinkedList()
lis.addRoot("0")
lis.insetUsingRec("0",0)
lis.insetUsingRec("1",2)
lis.insetUsingRec("2",3)
lis.insetUsingRec("3",4)
lis.insetUsingRec("3",5)
lis.print_list()
lis.removeDuplicates()
lis.print_list()
```

## Merge 2 sorted list

```python
class Node():
    def __init__(self, val=None):
        self.val = val
        self.next = None

    def __str__(self):
        return str(self.val)

class LinkList():
    def __init__(self):
        self.head = None
        self.size = 0
    
    def insertHead(self, val):
        self.head = Node(val)
        self.size += 1

    def insertEnd(self, val):
        nn = Node(val)
        temp = self.head
        if temp is not None:
            while temp.next is not None:
                temp = temp.next
            temp.next = nn
            self.size += 1
        else:
            self.head = Node(val)
            self.size += 1

    def print_list(self):
        this_node = self.head
        while this_node is not None:
            print(this_node,end='->')
            this_node = this_node.next
        print("None")

def mergeList(res:LinkList, lis1:LinkList, lis2:LinkList):
    head1 = lis1.head
    head2 = lis2.head
    while head1 is not None and head2 is not None:  
        if head1.val <= head2.val:
            res.insertEnd(head1.val)
            head1 = head1.next
        else:
            res.insertEnd(head2.val)
            head2 = head2.next

    while head2 is not None:
        res.insertEnd(head2.val)
        head2 = head2.next  
    while head1 is not None:
        res.insertEnd(head1.val)
        head1 = head1.next
    return res
            
lis1 = LinkList()
lis1.insertEnd(1)
lis1.insertEnd(2)
lis1.insertEnd(5)
lis1.print_list()

lis2 = LinkList()
lis2.insertEnd(1)
lis2.insertEnd(3)
lis2.insertEnd(4)
lis2.insertEnd(6)
lis2.insertEnd(8)
lis2.print_list()

lis3 = LinkList()
lis3 = mergeList(lis3, lis1, lis2)
lis3.print_list()
```

```python
class Node():
    def __init__(self, val=None):
        self.val = val
        self.next = None

    def __str__(self):
        return str(self.val)

class LinkList():
    def __init__(self, val=0):
        self.head = None if val==0 else Node(val)
        self.size = 0 if val==0 else 1

    def insertEnd(self, val):
        nn = Node(val)
        temp = self.head
        if temp is not None:
            while temp.next is not None:
                temp = temp.next
            temp.next = nn
            self.size += 1
        else:
            self.head = Node(val)
            self.size += 1
    
    def print_list(self):
        this_node = self.head
        print("Size: ", self.size)
        while this_node is not None:
            print(this_node,end='->')
            this_node = this_node.next
        print("None")

    def middle(self):
        num = (1)*self.size//(2)
        temp = self.head
        i=0
        while temp:
            if i != num:
                temp = temp.next
                i += 1
            else:
                return temp.val
    
    #floyd tortoise and hare (simlified)
    def middleBetter(self):
        slow = fast = self.head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        return slow.val

lis1 = LinkList()
lis1.insertEnd(5)
lis1.insertEnd(6)
lis1.insertEnd(7)
lis1.insertEnd(8)
#lis1.insertEnd(9)
lis1.print_list()
print(lis1.middle())
print(lis1.middleBetter())
```

```python
class Node():
    def __init__(self, val=None):
        self.val = val
        self.next = None

    def __str__(self):
        return str(self.val)

class LinkList():
    def __init__(self, val=0):
        self.head = None if val==0 else Node(val)
        self.size = 0 if val==0 else 1

    def insertEnd(self, val):
        nn = Node(val)
        temp = self.head
        if temp is not None:
            while temp.next is not None:
                temp = temp.next
            temp.next = nn
            self.size += 1
        else:
            self.head = Node(val)
            self.size += 1
    
    def print_list(self):
        this_node = self.head
        print("Size: ", self.size)
        while this_node is not None:
            print(this_node,end='->')
            this_node = this_node.next
        print("None")
    
    def makeCycle(self, index):
        if index == -1:
            return
        t = self.head
        while index:
            index -= 1
            t = t.next
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = t

def hasCycle(arr) -> bool:
    temp = arr.head
    dic = set()
    bf = False
    while temp:
        if temp in dic:
            bf = True
            break
        dic.add(temp)
        temp = temp.next
    return bf

def hasCycleTortAndHare(arr):
    slow = fast = arr.head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    return False

lis1 = LinkList()
arr1 = [1,2,3,4,3,2]
#arr1 = [-21,10,17,8,4,26,5,35,33,-7,-16,27,-12,6,29,-12,5,9,20,14,14,2,13,-24,21,23,-21,5]
pos = 2
for i in arr1:
    lis1.insertEnd(i)
lis1.makeCycle(pos)
#print(hasCycle(lis1))
print(hasCycleTortAndHare(lis1))
```

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def hasCycle(self, head) -> bool:
        slow = fast = head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                return True
        return False
        
```

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def cyc(self,head):
        slow = fast = head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                return slow
        return None
                
    def detectCycle(self, head):
        cyc = self.cyc(head)
        if cyc is None:
            return None
        else:
            p1 = head
            p2 = cyc
            while p1!=p2:
                p1 = p1.next
                p2 = p2.next
            return p1
```
