```python
def flippingMatrix(matrix):
    n = len(matrix[0])
    m = 0
    for i in range(int(n/2)):
        for j in range(int(n/2)):
            lis = (matrix[i][j], matrix[i][n-1-j], matrix[n-1-i][j], matrix[n-1-i][n-1-j])
            m += max(lis)
    return m
```

```python
class Node:
    def __init__(self,val = None):
        self.val = val
        self.next = None
    
class SLinkList:
    def __init__(self):
        self.head = None

    def listprint(self):
        printval = self.head
        while printval is not None:
            print(printval.val,end=' ')
            printval = printval.next
    
    def insertFront(self,newData):
        nn = Node(newData)
        nn.next = self.head
        self.head = nn
    
    def insertEnd(self,newData):
        nn = Node(newData)
        travnode = self.head
        while travnode.next is not None:
            travnode = travnode.next
        travnode.next = nn
    
    def insertAfter(self, preNode, newData):
        nn = Node(newData)
        travNode = self.head
        while travNode.val == preNode.val:
            travNode = travNode.next
        travNode = travNode.next
        nn.next = travNode.next
        travNode.next = nn
    
    def insertBefore(self, preNode, newData):
        nn = Node(newData)
        travNode = self.head
        while travNode.val == preNode.val:
            travNode = travNode.next
        nn.next = travNode.next
        travNode.next = nn

lis = SLinkList()
lis.head = Node("Mon")
e2 = Node("Tue")
e3 = Node("Wed")

lis.head.next = e2
e2.next = e3

#printing
lis.listprint()

#insert Front
lis.insertFront("Sun")
print("\nAfter insertion at front:")
lis.listprint()

#insert End
lis.insertEnd("Thu")
print("\nAfter insertion at end:")
lis.listprint()

#insert After
lis.insertAfter(lis.head.next, "Glob")
print("\nAfter insertion after an element:")
lis.listprint()

#insert Before
lis.insertBefore(lis.head.next, "Globy")
print("\nAfter insertion before an element:")
lis.listprint()
```
